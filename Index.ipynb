```python
#Calling the necessary classes 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression 
```

```python
#Storing the data
data = pd.read_csv("http://bit.ly/w-data")
```

```python
#Scatter plot
plt.scatter(x = "Hours",y = "Scores",data = data)
plt.title("Scatter Plot of Hours of study vs the Scores obtained")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

```python
#Extracting the data for correlation coefficient
Hours = data.iloc[:,0]  
Scores = data.iloc[:,1]
```

```python
np.corrcoef(Hours,Scores)[0,1]   #The correlation coefficient.
```

```python
#Thus we may find the best fitted straight line based on the data set.
X = Hours.values.reshape(-1,1)
y = Scores.values.reshape(-1,1)
model = LinearRegression().fit(X, y)
fit = model.predict(X)
model.score(X, y)   #R- squared, which is very high. 
b = model.coef_[0,0]   #Slope
a = model.intercept_[0] #Intercept
print("The estimated parameters are,",a,b)      #estimated values.
```

```python
#The Fitted model.
sns.regplot(x = Hours,y = Scores)
plt.title("The fitted regression line")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

```python
#Thus, the predicted value is,
print("Therefore, the predicted value of the score when the study time is 9.25 hrs per day is",model.predict(np.array([[9.25]]))[0,0])
```
