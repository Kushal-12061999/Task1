#Considering the necessary classes

```python
import tkinter.filedialog as tk
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
```

#Storing the data set

```python
data = pd.read_csv("Iris.csv")
data.head()
```

#Extracting the different species

```python
classes = data.Species.unique()
```

#Extracting the features 

```python
lengths = data.loc[:, ["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
```

#K means clustering to find the optimum number of clusters. Here the ELBOW RULE have been used.

```python
kmeans = KMeans(n_clusters = 99)               #initializing the number of clusters
wcss = []
for k in range(1,15):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(lengths)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,15),wcss,"-o")
plt.grid(True)
plt.title("Intertia vs Amount of clusters")
plt.xlabel("Amount of clusters")
plt.ylabel("Inertia")
plt.show()
```

#Thus, 3 seems to be the most optimal number of clusters.

#Lets check it visually.

```python
#For Petal length and width
plt.suptitle("Comparing a K- mean cluster(K = 3), by petal dimensions with the origina data.")
plt.subplot(1,2,1) 
kmeans = KMeans(n_clusters = 3)
lengths["labels"] = kmeans.fit_predict(lengths)
plt.scatter(lengths.PetalLengthCm[lengths.labels == 0],lengths.PetalWidthCm[lengths.labels == 0])
plt.scatter(lengths.PetalLengthCm[lengths.labels == 1],lengths.PetalWidthCm[lengths.labels == 1])
plt.scatter(lengths.PetalLengthCm[lengths.labels == 2],lengths.PetalWidthCm[lengths.labels == 2])
plt.title("By Clusters")
plt.xlabel("Petal Length(cm)")
plt.ylabel("Petal Width(cm)")
plt.subplot(1,2,2)
plt.scatter(data.PetalLengthCm[data.Species == "Iris-setosa"],data.PetalWidthCm[data.Species == "Iris-setosa"])
plt.scatter(data.PetalLengthCm[data.Species == "Iris-versicolor"],data.PetalWidthCm[data.Species == "Iris-versicolor"])
plt.scatter(data.PetalLengthCm[data.Species == "Iris-virginica"],data.PetalWidthCm[data.Species == "Iris-virginica"])
plt.title("Original data")
plt.xlabel("Petal Length(cm)")
plt.ylabel("Petal Width(cm)")
```

```python
#For Sepal length and width 
plt.suptitle("Comparing a K- mean cluster(K = 3), by sepal dimensions with the origina data.")
plt.subplot(1,2,1) 
kmeans = KMeans(n_clusters = 3)
lengths["labels"] = kmeans.fit_predict(lengths)
plt.scatter(lengths.SepalLengthCm[lengths.labels == 0],lengths.SepalWidthCm[lengths.labels == 0])
plt.scatter(lengths.SepalLengthCm[lengths.labels == 1],lengths.SepalWidthCm[lengths.labels == 1])
plt.scatter(lengths.SepalLengthCm[lengths.labels == 2],lengths.SepalWidthCm[lengths.labels == 2])
plt.title("By Clusters")
plt.xlabel("Sepal Length(cm)")
plt.ylabel("Sepal Width(cm)")
plt.subplot(1,2,2)
plt.scatter(data.SepalLengthCm[data.Species == "Iris-setosa"],data.SepalWidthCm[data.Species == "Iris-setosa"])
plt.scatter(data.SepalLengthCm[data.Species == "Iris-versicolor"],data.SepalWidthCm[data.Species == "Iris-versicolor"])
plt.scatter(data.SepalLengthCm[data.Species == "Iris-virginica"],data.SepalWidthCm[data.Species == "Iris-virginica"])
plt.title("Original data")
plt.xlabel("Sepal Length(cm)")
plt.ylabel("Sepal Width(cm)")
```
